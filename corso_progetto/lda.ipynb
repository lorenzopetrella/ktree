```python
from custom_libraries.miscellaneous import progress_bar
from custom_libraries.image_dataset import *
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
weighting = 'paired'
trials = 10

classes = np.load(project_folder + 'results/classes.npy', allow_pickle=True)
score_test = np.zeros((len(classes), trials))

for j, (t1, t2, ds) in enumerate(classes):

    print(f"- Dataset: {ds}")

    test_ds = ImageDataset(ds, 'test', data_dir=None, USPS_dir=project_folder + 'USPS/')
    train_ds = ImageDataset(ds, 'train', data_dir=None, USPS_dir=project_folder + 'USPS/')

    for x in [train_ds, test_ds]:
        x.filter(t1, t2, overwrite=True)
        x.shuffle()
        x.normalize()
        if x.images.shape[1:3] == (28, 28):
            x.pad()
        x.vectorize(True)

    X_test, y_test = test_ds.subset(validation=False)
    X_train, y_train, _, _ = train_ds.subset(shard=True, shard_number=trials, validation=True, validation_size=len(X_test))

    average_acc = 0

    for m in range(trials):
        lda = LinearDiscriminantAnalysis()
        lda.fit(X_train[m], y_train[m])
        score_test[j, m] = lda.score(X_test, y_test)
        average_acc = average_acc + score_test[j, m]

        # Save accuracy array
        np.save(project_folder + 'results/lda_score_test.npy', score_test)

    progress_bar(10, max=10, pre=f"-- Pair {t1} - {t2}:\t", post=f"\tAccuracy = {average_acc / trials}")
    print()
```
